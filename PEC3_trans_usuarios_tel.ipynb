```python
import pandas as pd
import openpyxl as op
import numpy as np
```

```python
df1 = pd.read_csv('number-of-internet-users-by-country.csv')
df1
```

```python
df1 = df1.rename(columns={'Number of internet users (OWID based on WB & UN)':'Internet_users'})
df1
```

```python
df1.columns.values
```

```python
np.unique(df1['Entity'])
```

```python
df1.drop(df1[df1["Entity"] == "World"].index, inplace=True)
```

```python
df1 = pd.pivot_table(df1, index=['Entity'], columns = ['Year'], values = ['Internet_users'], aggfunc = 'sum')
```
